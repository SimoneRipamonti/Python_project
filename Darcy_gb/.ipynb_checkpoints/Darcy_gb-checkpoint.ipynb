```python
import numpy as np
import scipy.sparse as sps
import porepy as pp
import math
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append("../class")
from Flow_class import Flow
```

```python
Nx=20
Ny=20
```

```python
gb=pp.meshing.cart_grid(fracs=[],nx=[Nx,Ny],physdims=[1,1])
domain={"xmin":0.0,
       "xmax":1.0,
       "ymin":0.0,
       "ymax":1.0}
```

```python
pp.plot_grid(gb)
```

```python
parameter={"aperture":1e-3,
          "fracture_perm":1e-5,
          "perm":1,
          "f_lambda":lambda x,y,z:math.pi*math.pi*2*math.sin(math.pi*x)*math.sin(math.pi*y),
          "bc_value":[0.0,0.0],
          "bc_type":['dir','dir'],
          "method":"Tpfa"}
```

```python
flow=Flow(gb,domain,parameter)
```

```python
flow.set_data()
flow.discretize()
flow.solve()
flow.plot_pressure()
```
